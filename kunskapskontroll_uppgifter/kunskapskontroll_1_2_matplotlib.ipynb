### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

**Answer:**

The two ways of plotting in Matplotlib are:
1. Object Oriented Style:
    Explicitly creating figures and axes and calling on them.
2. Pyplot Style:
    Let Pyplot implicitly create and manage the figures and axes.

The recommended approach is the object-oriented style, especially for more complicated plots and for functions and scripts that are intended to be reused in larger projects.
Pyplot-style can be used to get a quick view.

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

**Answer:**

Figure: An object containing all the plot elements. 
Axes:   Represents one plot in a figure. Contains the plotted data, axis, ticks, labels etc. A figure can have multiple axes. 
Axis:   Defines the coordinate system for the data in Axes. An axes typically has two Axis objects.
        Axis are used to add annotations to the plot, like labels for x- and y-axis, titles and legends.

Artist: The Base Class for almost all plot objects. 

### When plotting in Matplotlib, what is the expected input data type?

**Answer:**

Plotting functions expect numpy.array or numpy.ma.masked.array as input, or objects that can be passed to numpy.asarray.

Matplotlib can also generate plots from string-indexable objects like a dict or a pandas.DataFrame. This can be done through the [data] keyword, by passing the strings corresponding to the x and y variables.

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
x = np.linspace(-4, 4)
y = x**2

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("Function y = x^2")
ax.set_xlabel("x-axis")
ax.set_ylabel("y-axis")
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {"grapes": 22, "apple": 8, "orange": 15, "lemon": 20, "lime": 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].scatter(x, y)
axs[0].set_title("y = x^2")
axs[0].set_xlabel("x-axis")
axs[0].set_ylabel("y-axis")
axs[1].bar(names, values)
axs[1].set_title("Fruit data")
axs[1].set_xlabel("Name")
axs[1].set_ylabel("Value")
```
