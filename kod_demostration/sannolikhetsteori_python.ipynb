## Ilustrate the Law of Large Numbers (LLN)  [Called "De relative frekvensernas stabilitet" in the book p.14]
1. Simulate 10000 dice throws. 
2. Plot the relative frequency of the number of times three dots occur for 1, 2, ..., 10000 throws.
3. To what number does the relative frequency converge to? What is the probability of getting three dots when throwing a dice? Do these two numbers coincide?
4. Explain what you observe. 

Make sure to create a good pedagogical plot, i.e. include title, x-label and y-label. 

**Hint: For step 1 use the random.choices() function. You should also create an empty list that you append with the relative frequencies for 1, 2, ... 10000 throws.**

```python
# Help about the random.choices function
help(random.choices)

print(random.choices([1, 2, 3, 4, 5, 6,], k=3))
print(random.choices([1, 2, 3, 4, 5, 6,], k=10))
```

```python
three_dots = []
relative_freq = []

for i in range(1,10001):
    
    throws = np.array(random.choices([1, 2, 3, 4, 5, 6,], k = 1))
    three_dots.append(throws == 3)
    relative_freq.append(np.mean(three_dots))

fig, ax = plt.subplots()
ax.plot(relative_freq)
ax.axhline(1/6, color = 'g')
ax.set_title('Dice roll - LLN')
ax.set_ylabel('Relative Frequency')
ax.set_xlabel('Dice throws')
```

**3. To what number does the relative frequency converge to? What is the probability of getting three dots when throwing a dice? Do these two numbers coincide?**

**Answer:**\
Yes. The relative frequency converges around 1/6. Which is the same probability of getting three dots on a dice, as there are 6 equal outcomes when rolling a die.

**4. Explain what you observe**

The law of large numbers states that the average of results obtained from a large number of independent random samples converges to the true average value.\
This is confirmed as the average is much higher in the first 1000 or so samples. But is gradually converging around the true average.

**Simulate 200 coin tosses and study the outcome of heads. Create a plot that displays the frequency.**

```python
def coin_toss():
    heads = []
    relative_freq = []
    for i in range(1, 201):
        flip = np.array(random.choices(["heads", "tails"], k=1))
        heads.append(flip == "heads")
        relative_freq.append(np.mean(heads))
    return relative_freq


simulation = []
simulation.append(coin_toss())
simulation.append(coin_toss())

fig, ax = plt.subplots(nrows=1, ncols=2)
fig.set_figwidth(10)
fig.set_figheight(4)
ax[0].plot(simulation[0])
ax[0].axhline(1 / 2, color="g")
ax[0].set_title("Frequency of heads - LLN")
ax[0].set_ylabel("Relative Frequency")
ax[0].set_xlabel("Coin tosses")

ax[1].plot(simulation[1])
ax[1].axhline(1 / 2, color="g")
ax[1].set_title("Frequency of heads - LLN")
ax[1].set_ylabel("Relative Frequency")
ax[1].set_xlabel("Coin tosses")
```

**Answer**\
a) The probability of getting 'heads' in a coin toss is 50% indicated by the green line, which we can gradually see.  
b) Vad menas med de relativa frekvensernas stabilitet?\

# Calculate 1!, 2!, ..., 12! . What is your conclusion regarding the growth of the factorial function?

```python
n = []
factorial = []
for i in range(1, 4):
    n.append(i)
    factorial.append(np.prod(n))

fig, ax = plt.subplots()
ax.plot(factorial)
ax.set_title("Factorial growth")
ax.set_ylabel("Relative Frequency")
ax.set_xlabel("Dice throws")
ax.set_xticks(ticks=np.arange(0, max(n)))
```
